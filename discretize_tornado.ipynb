```python
import numpy as np
from matplotlib import pyplot as plt
from utils import *
```

```python
# parameters
turns = 3
ms_per_revolution = 9600
total_time = 1200 # sec
vi = 48 # ms/sec
angle_step = 10 # deg

# calculate implicit parameters
total_ms = turns * ms_per_revolution
ms_step = 9600/360*angle_step
steps = int(round(total_ms / ms_step + 1))
print("Total steps: {}".format(steps))

# calculate acceleration and then time over for discrete positions
acceleration = (total_ms - vi*total_time)*2/total_time**2
print(acceleration)

table = np.ndarray((int(steps), 2))
#print(np.linspace(0, total_ms, ms_step))
for i in range(0, steps):
    disc = vi**2 + 4*(acceleration/2)*i*ms_step
    if disc > 0:
        time = (-vi + np.sqrt(vi**2 + 4*(acceleration/2)*i*ms_step))/acceleration
    else: 
        time = total_time
    table[i,:] = (i*ms_step, time)

print("{}".format(table))
print("{}".format(table[:-1,1] - table[1:,1]))

fig, ax = plt.subplots()
ax.plot(table[:,1], table[:,0])
#ax.plot(table[:-1,1] - table[1:,1])
plt.show()
```
